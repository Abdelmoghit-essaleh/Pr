```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,5,8,8.5])
plt.scatter(x,y,color='r')
```

```python
xm=np.mean(x)
ym=np.mean(y)
xn=xd=0
for i in range(len(x)):
    xn+=(x[i]-xm)*(y[i]-ym)
    xd+=(x[i]-xm)**2
    
a=xn/xd
b=ym-a*xm
yc=a*x+b
plt.scatter(x,y,color='b')
plt.plot(x,yc,'r')
```
