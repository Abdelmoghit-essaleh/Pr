```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x=np.array([200,250,130,70])
y=np.array([30,33,17.5,9.2])
plt.scatter(x,y,color='g')
```

```python
xm=np.mean(x)
ym=np.mean(y)
xn=xd=0
for i in range(len(x)):
    xn+=(x[i]-xm)*(y[i]-ym)
    xd+=(x[i]-xm)**2

a=xn/xd
b=ym-a*xm
yc=a*x+b
plt.scatter(x,y,color='g')
plt.plot(x,yc,'b')
    
```
